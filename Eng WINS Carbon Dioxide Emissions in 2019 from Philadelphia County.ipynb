```python
import folium
import pandas as pd
#from palettable.colorbrewer.sequential import RdPu_8#import branca
#m = folium.Map(location=[39.952583, -75.165222], tiles="OpenStreetMap", zoom_start=13)
#the data in carbondioxide.csv is sorted from greatest to least
data = pd.read_csv('carbondioxide.csv')
```

```python
m = folium.Map(location=[39.952583,  -75.165222], tiles="OpenStreetMap", zoom_start=13)
m.save("C:/Users/we are/Documents/GitHub/Drexel/Eng WINS Carbon Dioxide Emissions in 2019 from Philadelphia County.html")
for i in range(0,len(data)): 
    html=f"""
        <h1> {data.iloc[i]['Facility Name']}</h1>
        <p>Emission Amt in tons:</p>
        <ul>
            <li>{data.iloc[i]['Emission Amt(In Tons)']}</li>
        <ul>
        """
    iframe = folium.IFrame(html=html, width=300, height=300)
    popup = folium.Popup(iframe, max_width=2650)
    if i >= 0 and i < 18:  
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "darkred",
        fill = True,
        fill_color="darkred",
        fill_opacity=.5,
     ).add_to(m)
        
    if i >= 18 and i < 35:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "red",
        fill = True,
        fill_color="red",
        fill_opacity=.5,
     ).add_to(m) 
        
    if i >= 35 and i < 52:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "orange",
        fill = True,
        #fill_color="orange",
        #fill_opacity=1,
     ).add_to(m)
        
    if i >= 52 and i < 69:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "pink",
        fill = True,
        #fill_color="pink",
        #fill_opacity=1,
     ).add_to(m)
        
    if i >= 69 and i < 86:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "beige",
        fill = True,
        #fill_color="beige",
        #fill_opacity=1,
     ).add_to(m)
        
    if i >= 86 and i < 108:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 6,
        popup = popup,
        color = "gray",
        fill = True,
        #fill_color="gray",
        #fill_opacity=1,
     ).add_to(m)
            
    if i >= 108 and i < 120:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 5,
        popup = popup,
        color = "lightgray",
        fill = True,
        #fill_color="lightgray",
        #fill_opacity=1,
     ).add_to(m)
        
    if i >= 120 and i < 139:
        folium.CircleMarker(
        location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
        radius = 5,
        popup = popup,
        color = "black",
        fill = True,
        #fill_color="black",
        #fill_opacity=1,
     ).add_to(m)
m
```
